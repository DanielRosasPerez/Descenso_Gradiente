## **A**

```python
peso = 0.5
entrada = 0.5
salida_deseada = 0.8

for iteracion in range(20):
    prediccion = entrada * peso
    error = (prediccion - salida_deseada)**2
    gradiente = 2*(prediccion - salida_deseada)*entrada # Esto se obtiene aplicando la regla de la cadena.
    peso -= gradiente
    
    print(f"Error: {error} ; Predicción: {prediccion}")
```

## **B**

```python
peso, salida_deseada, entrada = (0.1, 0.8, 0.5)

for iteracion in range(1,55+1):
    
    prediccion = entrada * peso    
    error = (prediccion - salida_deseada)**2
    # Descenso del Gradiente:
    delta = 2*(prediccion - salida_deseada) # Error delta de la neurona de salida.
    peso_delta = delta * entrada # Que tanto influyo el peso en la neurona de salida.
    # Actualización del parámetro peso:
    peso -= peso_delta
    
    print("Error {} ; Predicción: {}".format(error, prediccion))
    
    if prediccion == salida_deseada:
        print(f"\nEn la iteración número {iteracion} se obtuvo la salida deseada\n")
        break
```

## **F**

Función de error vs pesos, haciendo uso del Error Cuadrático: 
$$ error(w1,w2) = ((-0.2*w1 + 0.4*w2) - 1)^2 $$

Función de error vs predicción, haciendo uso del Error Cuadrático:
$$ error(prediccion) = (prediccion - 1)^2 $$

*Primero, entrenamos nuestro algortimo:*

```python
def generador(x):
    """Esta función es un generador."""
    
    for i in range(x):
        yield i

def entrenamiento_red(entradas, pesos, salida_deseada, alfa, iteraciones):
    """Esta función tiene como propósito entrenar a nuestra red neuronal."""
    
    predicciones = list() # Guardará las predicciones hechas en cada iteración.
    errores = list() # Guardará los errores cometidos durante el entrenamiento.
    lista_pesos = list() # Guardará la evolución del entrenamiento de nuetros pesos.

    for iteraciones in generador(int(iteraciones)):

        prediccion = entradas@pesos
        predicciones.append(prediccion)

        error = (prediccion - salida_deseada)**2
        errores.append(error)

        # Gradiente:
        delta = 2*(prediccion - salida_deseada) # Error Delta en común.<
        pesos_delta = delta * entradas

        # Actualización parámetros:
        lista_pesos.append(np.array([pesos[0],pesos[1]]))
        pesos -= alfa*pesos_delta
        
        if prediccion == salida_deseada:
            
            # Con la finalidad de guardar la última actualización de los datos dado que en el útlimo ciclo,
            # los pesos se vuelven a actualizar y como ya no se entra al bucle los datos restantes no son calculados
            # en función de esa última actualización.
            predicciones.append(entradas@pesos)
            errores.append((prediccion - salida_deseada)**2)
            lista_pesos.append(np.array([pesos[0],pesos[1]]))
            
            print(f"\nEn la iteración número '{iteraciones+1}' se obtuvo el valor exacto de la salida deseada\n")
            
            return predicciones, errores, lista_pesos
            
    # Con la finalidad de guardar la última actualización de los datos dado que en el útlimo ciclo,
    # los pesos se vuelven a actualizar y como ya no se entra al bucle los datos restantes no son calculados
    # en función de esa última actualización.
    predicciones.append(entradas@pesos)
    errores.append((prediccion - salida_deseada)**2)
    lista_pesos.append(np.array([pesos[0],pesos[1]]))
            
    return predicciones, errores, lista_pesos
```

```python
# Declaramos nuestros datos:
entradas, pesos, salida_deseada = np.array([-0.2, 0.4]), np.array([1.2, 0.8]), 1.0
alfa = 0.9

# Entrenamos nuestro modelo:
predicciones, errores, lista_pesos = entrenamiento_red(entradas, pesos, salida_deseada, alfa, 102)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20,10))

grafica = 1

for i in range(axes.shape[0]):
    for j in range(axes.shape[1]):
        axes[i,j] = fig.add_subplot(int(str(axes.shape[0])+str(axes.shape[1])+str(grafica)),projection="3d")
        grafica += 1

#w1 = np.linspace(1,-6,num=1000)
#w2 = np.linspace(-1,3,num=1000)

w1 = np.linspace(-3,5,num=1000)
w2 = np.linspace(-1,8,num=1000)

def funcion_3D(peso1,peso2):
    
    funcion = ((-0.2*peso1 + 0.4*peso2)-1)**2
    
    return funcion

W1, W2 = np.meshgrid(w1,w2)
Z = funcion_3D(W1,W2)

index = [0,2,4,5,6,100] # Index de los pesos a tomar de la lista "lista_pesos".
contador = 0

for i in range(axes.shape[0]):
    
    for j in range(axes.shape[1]):

        # Gráficamos nuestra función de error (error(w1,w2)):
        axes[i,j].plot_surface(W1,W2,Z, cmap="winter", alpha=0.4)
        
        axes[i,j].set_xlabel("w1", fontsize=15, color="blue")
        axes[i,j].set_ylabel("w2", fontsize=15, color="blue")
        axes[i,j].set_zlabel("Error(w1,w2)", fontsize=15, color="blue")

        # Indicamos el mínimo global de la función mediante un punto azul:
        axes[i,j].plot([1.2],[3.1],[0], color="blue", marker='o', ms=15, label="Mínimo Global (1.2,3.1,0)")

        # Gráficamos el ajuste de los pesos:
        coordenada_w1 = lista_pesos[index[contador]][0]
        coordenada_w2 = lista_pesos[index[contador]][1]
        coordenada_error = errores[index[contador]]
        
        etiqueta = '(' + str(np.around(coordenada_w1,3)) + ',' + str(np.around(coordenada_w2,3)) + ',' + str(np.around(coordenada_error,3)) + ')'
        axes[i,j].plot([coordenada_w1], # Peso 1.
                     [coordenada_w2], # Peso 2.
                     [coordenada_error], # Función error(w1,w2). 
                     color="red", marker='o', ms=15, label="Pesos " + etiqueta) # configuración del punto.
        
        axes[i,j].legend(fontsize=13) # Habilitamos la impresión de etiquetas (labels) en las gráficas y ajustamos el tamaño de letra de las mismas.
        
        # Declaramos el formato de las coordenadas:
        axes[i,j].text(0, 7, 7, "Formato Coordenadas:\n(w1,w2,error(w1,w2))", color="darkslateblue", fontsize=15)
        
        # Declaramos el respectivo número de iteración en cada gráfica:
        axes[i,j].text(0, 6, 5, "Iteración # " + str(index[contador]), color="green", fontsize=15)

        # Ajustamos el ángulo de percepción de la gráfica:
        axes[i,j].view_init(25,180) # (45,0)
        
        contador += 1

plt.tight_layout()
plt.show()
```

Como podemos observar, conforme nuestra red neuronal va entrenando, nuestros pesos (punto rojo) comienzan a acercarse al mínimo de la función $error(w1,w2)$ hata que finalmente lo alcanzan.  

**Cabe destacar, que el mínimo global de la función esta representado por el punto azul. Por otro lado, cualquier valor que haga que la función $error(w1,w2)$ sea igual a 0, será un valor valído para nuestros pesos.**

### **NOTA**  
El error de esta neurona (3 entradas, 3 pesos y 1 salida), se encuentra en función de los 3 pesos y esta dado por:  

$$ error(w1,w2,w3) = ((entrada_1*w1 + entrada_2*w2 + entrada_3*w3) - 1)^2 $$

**Recordemos que las ENTRADAS son FIJAS, lo único que VARIAMOS son los PESOS**. Es por eso que $error(w1,w2,w3) \neq (prediccion - 1)^2$ La primer función se encuentra en 4D dado que tenemos 3 variables que al sustituirse todas arrojan un 1 resultado. Por otro lado, la segunda función, es una función cuya gráfica esta en 2D dado que depende de una variable que al sustuirse por cualquier número real arroja 1 resultado.

Sin embargo, si yo intento separar cada neurona con su respectivo peso y compararlo directamente con la función de error, es decir:

$$error_{n1}(w1) = ((entrada_1*w1)-1)^2$$
$$error_{n2}(w2) = ((entrada_2*w2)-1)^2$$
$$error_{n3}(w3) = ((entrada_2*w3)-1)^2$$  

Al gráficar estas funciones (error vs pesos) y obtener el valor mínimo para cada una (por ende, el valor del peso en el eje x), y sustituirla en $error(w1,w2,w3)$ nos percatamos que el error no es 0 ni siquiera se aproxima, además la predicción es 4. Por lo tanto, yo concluyo que no existe relación entre separar cada neurona y calcular su mínimo correspondiente usando la misma fórmula para el error de la nuerona de salida.

```python
def generador(x):
    """Está función es un generador."""
    
    for i in range(x):
        yield i

def algoritmo_entrenamiento(entradas, pesos, salida_deseada, alfa, epocas):
    """Está función lleva a cabo el entrenamiento de nuestra red."""
    
    predicciones = list() # Guardará las predicciones hechas en cada iteración.
    errores = list() # Guardará los errores cometidos durante el entrenamiento.
    lista_pesos = list() # Guardará la evolución del entrenamiento de nuetros pesos.

    for iteraciones in generador(int(epocas)+1):

        prediccion = entradas@pesos
        predicciones.append(prediccion)

        error = (prediccion - salida_deseada)**2
        errores.append(error)

        # Gradiente:
        delta = 2*(prediccion - salida_deseada) # Error Delta en común.
        pesos_delta = delta * entradas

        # Actualización parámetros:
        lista_pesos.append(pesos)
        pesos -= alfa*pesos_delta
        
        if (np.around(prediccion,10) == salida_deseada).sum() == len(prediccion): # Valor aproximado.
            
            # Con la finalidad de guardar la última actualización de los datos dado que en el útlimo ciclo,
            # los pesos se vuelven a actualizar y como ya no se entra al bucle los datos restantes no son calculados
            # en función de esa última actualización.
            predicciones.append(entradas@pesos)
            errores.append((prediccion - salida_deseada)**2)
            lista_pesos.append(pesos)
            
            if (prediccion == salida_deseada).sum() == len(prediccion): # Valor exacto.
                
                print(f"\nEn la iteración número '{iteraciones+1}' se obtuvo el VALOR EXACTO de la salida deseada\n")
                break
            
            print(f"\nEn la iteración número '{iteraciones+1}' se obtuvo un VALOR MUY APROXIMADO a la salida deseada\n")
            break
            
    return predicciones, errores, lista_pesos
```

```python
def generador (x):
    """Esta función es un generador."""
    
    for i in range(x):
        
        yield i
        
def algoritmo_de_entrenamiento(entradas, pesos, salidas_deseadas, alfa, epocas):
    """Esta función ejecuta el algoritmo de entrenamiento de nuestra red."""
    
    predicciones = list() # Guardará las predicciones hechas en cada iteración.
    errores = list() # Guardará los errores cometidos durante el entrenamiento.
    lista_pesos = list() # Guardará la evolución del entrenamiento de nuetros pesos.

    for iteraciones in generador(epocas+1):

        prediccion = entradas@pesos
        predicciones.append(prediccion)

        error = (prediccion - salidas_deseadas)**2
        errores.append(error)

        # Gradiente:
        deltas = 2*(prediccion - salidas_deseadas)*1 # Error Delta en común.
        pesos_delta = np.asarray([[delta*entrada for entrada in entradas] for delta in deltas])

        # Actualización parámetros:
        lista_pesos.append(pesos)
        pesos -= alfa*pesos_delta.T
        
        if (np.around(prediccion,15) == salidas_deseadas).sum() == len(prediccion): # Valor aproximado.
            
            # Con la finalidad de guardar la última actualización de los datos dado que en el útlimo ciclo,
            # los pesos se vuelven a actualizar y como ya no se entra al bucle los datos restantes no son calculados
            # en función de esa última actualización.
            predicciones.append(entradas@pesos)
            errores.append((prediccion - salidas_deseadas)**2)
            lista_pesos.append(pesos)
            
            if (prediccion == salidas_deseadas).sum() == len(prediccion): # Valor exacto.
                
                print(f"\nEn la iteración número '{iteraciones+1}' se obtuvo el VALOR EXACTO de la salida deseada\n")
                break
            
            print(f"\nEn la iteración número '{iteraciones+1}' se obtuvo un VALOR MUY APROXIMADO a la salida deseada\n")
            break
            
    return predicciones, errores, lista_pesos
```
